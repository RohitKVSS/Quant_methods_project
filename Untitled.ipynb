```python
import numpy as np
import matplotlib.pyplot as plt
from math import log, sqrt, exp
from scipy.stats import norm


S0    = 458.96
K     = 460.0
T     = 70/252
r     = 0.035
q     = 0.0
sigma = 0.533
```

```python
def binomial_option_crr(S0, K, T, r, sigma, N, option_type='call', american=True, q=0.0):
    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1.0 / u
    disc = np.exp(-r * dt)
    p = (np.exp((r - q) * dt) - d) / (u - d)

    if not (0.0 <= p <= 1.0):
        raise ValueError(f"Risk-neutral probability out of bounds: p={p:.6f}")

 
    j = np.arange(N + 1)
    ST = S0 * (u ** j) * (d ** (N - j))

   
    if option_type == 'call':
        V = np.maximum(ST - K, 0.0)
    elif option_type == 'put':
        V = np.maximum(K - ST, 0.0)
    else:
        raise ValueError("option_type must be 'call' or 'put'")

   
    for i in range(N - 1, -1, -1):
        V_cont = disc * (p * V[1:i+2] + (1 - p) * V[0:i+1])

        if american:
            j = np.arange(i + 1)
            S_i = S0 * (u ** j) * (d ** (i - j))
            if option_type == 'call':
                exercise = np.maximum(S_i - K, 0.0)
            else:
                exercise = np.maximum(K - S_i, 0.0)
            V = np.maximum(V_cont, exercise)
        else:
            V = V_cont

    return float(V[0])
```

```python
N0 = 70
amer_call = binomial_option_crr(S0,K,T,r,sigma,N0,'call',american=True,q=q)
amer_put  = binomial_option_crr(S0,K,T,r,0.5095,N0,'put', american=True,q=q)
print("American Call (binomial):", amer_call)
print("American Put  (binomial):", amer_put)
```

```python
def bs_price_greeks(S0, K, T, r, sigma, option_type='call', q=0.0):
    sqrtT = np.sqrt(T)
    d1 = (np.log(S0 / K) + (r - q + 0.5 * sigma**2) * T) / (sigma * sqrtT)
    d2 = d1 - sigma * sqrtT

    Nd1, Nd2 = norm.cdf(d1), norm.cdf(d2)
    pdf_d1 = norm.pdf(d1)

    disc_r = np.exp(-r * T)
    disc_q = np.exp(-q * T)

    if option_type == 'call':
        price = S0 * disc_q * Nd1 - K * disc_r * Nd2
        delta = disc_q * Nd1
        theta = (-S0 * disc_q * pdf_d1 * sigma / (2 * sqrtT)
                 + q * S0 * disc_q * Nd1
                 - r * K * disc_r * Nd2)
    else:
        price = K * disc_r * norm.cdf(-d2) - S0 * disc_q * norm.cdf(-d1)
        delta = disc_q * (Nd1 - 1)
        theta = (-S0 * disc_q * pdf_d1 * sigma / (2 * sqrtT)
                 - q * S0 * disc_q * norm.cdf(-d1)
                 + r * K * disc_r * norm.cdf(-d2))

    gamma = disc_q * pdf_d1 / (S0 * sigma * sqrtT)
    vega  = S0 * disc_q * pdf_d1 * sqrtT
    return float(price), float(delta), float(gamma), float(theta), float(vega)
```

```python
bs_call_price, *_ = bs_price_greeks(S0,K,T,r,sigma,'call',q=q)
bs_put_price,  *_ = bs_price_greeks(S0,K,T,r,0.5092,'put', q=q)
print("BS European Call:", bs_call_price)
print("BS European Put :", bs_put_price)
```

```python
Ns = [5, 10, 20, 50, 75, 100, 150, 200, 300, 500, 750, 1000]
```

```python
bin_call = []
err_call = []
for N in Ns:
    pN = binomial_option_crr(S0,K,T,r,sigma,N,'call',american=True,q=q)
    bin_call.append(pN)
    err_call.append(pN - bs_call_price)

plt.figure(figsize=(7,4))
plt.plot(Ns, bin_call, marker='o', label='American Call (Binomial)')
plt.axhline(bs_call_price, linestyle='--', label='BS (European Call)')
plt.xlabel("N"); plt.ylabel("Price")
plt.title("Convergence: American Call Binomial  BlackScholes (q=0)")
plt.grid(True); plt.legend(); plt.show()

plt.figure(figsize=(7,4))
plt.plot(Ns, np.abs(err_call), marker='o')
plt.yscale('log')
plt.xlabel("N"); plt.ylabel("|Binomial  BS| (log)")
plt.title("Convergence Error: Call")
plt.grid(True, which="both", ls="--"); plt.show()
```

```python
bin_put = []
err_put_vs_bs = []
for N in Ns:
    pN = binomial_option_crr(S0,K,T,r,sigma,N,'put',american=True,q=q)
    bin_put.append(pN)
    err_put_vs_bs.append(pN - bs_put_price)

plt.figure(figsize=(7,4))
plt.plot(Ns, bin_put, marker='o', label='American Put (Binomial)')
plt.axhline(bs_put_price, linestyle='--', label='BS (European Put)')
plt.xlabel("N"); plt.ylabel("Price")
plt.title("American Put (Binomial) vs BlackScholes European Put")
plt.grid(True); plt.legend(); plt.show()


N_ref = 3000
ref_put = binomial_option_crr(S0,K,T,r,sigma,N_ref,'put',american=True,q=q)
err_put_ref = [p - ref_put for p in bin_put]

plt.figure(figsize=(7,4))
plt.plot(Ns, np.abs(err_put_ref), marker='o')
plt.yscale('log')
plt.xlabel("N"); plt.ylabel("|Binomial  Ref| (log)")
plt.title(f"Convergence Error: American Put (reference N={N_ref})")
plt.grid(True, which="both", ls="--"); plt.show()
```

```python
def binomial_greeks_fd(S0, K, T, r, sigma, N, option_type='put', q=0.0,
                       dS=1.0, dSigma=0.01, dT=1/252):
    V0 = binomial_option_crr(S0,K,T,r,sigma,N,option_type=option_type,american=True,q=q)

   
    V_up = binomial_option_crr(S0+dS,K,T,r,sigma,N,option_type=option_type,american=True,q=q)
    V_dn = binomial_option_crr(S0-dS,K,T,r,sigma,N,option_type=option_type,american=True,q=q)
    delta = (V_up - V_dn) / (2*dS)
    gamma = (V_up - 2*V0 + V_dn) / (dS**2)

  
    V_su = binomial_option_crr(S0,K,T,r,sigma+dSigma,N,option_type=option_type,american=True,q=q)
    V_sd = binomial_option_crr(S0,K,T,r,sigma-dSigma,N,option_type=option_type,american=True,q=q)
    vega = (V_su - V_sd) / (2*dSigma)

    
    T_dn = max(T - dT, 1e-6)
    V_Tdn = binomial_option_crr(S0,K,T_dn,r,sigma,N,option_type=option_type,american=True,q=q)
    theta_time_decay = (V_Tdn - V0) / dT   # negative typically
    return float(V0), float(delta), float(gamma), float(theta_time_decay), float(vega)
```

```python
N_sens = 500
```

```python
def sensitivity_plots_american(option_type):
   
    S_grid = np.linspace(0.7*S0, 1.3*S0, 30)
    opt, delt, gam = [], [], []
    for S in S_grid:
        V, D, G, Th, Ve = binomial_greeks_fd(S, K, T, r, sigma, N_sens, option_type=option_type, q=q)
        opt.append(V); delt.append(D); gam.append(G)

    plt.figure(figsize=(7,4)); plt.plot(S_grid, opt)
    plt.xlabel("Spot (S0)"); plt.ylabel(f"{option_type.capitalize()} Price")
    plt.title(f"American {option_type.capitalize()} Price vs Spot (Binomial)")
    plt.grid(True); plt.show()

    plt.figure(figsize=(7,4)); plt.plot(S_grid, delt)
    plt.xlabel("Spot (S0)"); plt.ylabel("Delta")
    plt.title(f"Delta vs Spot (American {option_type.capitalize()}, Binomial)")
    plt.grid(True); plt.show()


 
    sig_grid = np.linspace(0.2, 1.0, 30)
    opt, veg = [], []
    for sig in sig_grid:
        V, D, G, Th, Ve = binomial_greeks_fd(S0, K, T, r, sig, N_sens, option_type=option_type, q=q)
        opt.append(V); veg.append(Ve)

    plt.figure(figsize=(7,4)); plt.plot(sig_grid, opt)
    plt.xlabel("Volatility (sigma)"); plt.ylabel(f"{option_type.capitalize()} Price")
    plt.title(f"American {option_type.capitalize()} Price vs Volatility (Binomial)")
    plt.grid(True); plt.show()

    plt.figure(figsize=(7,4)); plt.plot(sig_grid, veg)
    plt.xlabel("Volatility (sigma)"); plt.ylabel("Vega")
    plt.title(f"Vega vs Volatility (American {option_type.capitalize()}, Binomial)")
    plt.grid(True); plt.show()

  
    T_grid = np.linspace(10/252, 200/252, 30)
    opt, thet, veg = [], [], []
    for TT in T_grid:
        V, D, G, Th, Ve = binomial_greeks_fd(S0, K, TT, r, sigma, N_sens, option_type=option_type, q=q)
        opt.append(V); thet.append(Th); veg.append(Ve)

    plt.figure(figsize=(7,4)); plt.plot(T_grid, opt)
    plt.xlabel("T (years)"); plt.ylabel(f"{option_type.capitalize()} Price")
    plt.title(f"American {option_type.capitalize()} Price vs Time to Maturity (Binomial)")
    plt.grid(True); plt.show()

    plt.figure(figsize=(7,4)); plt.plot(T_grid, thet)
    plt.xlabel("T (years)"); plt.ylabel("Theta (time decay per year)")
    plt.title(f"Theta vs Time to Maturity (American {option_type.capitalize()}, Binomial)")
    plt.grid(True); plt.show()

    plt.figure(figsize=(7,4)); plt.plot(T_grid, veg)
    plt.xlabel("T (years)"); plt.ylabel("Vega")
    plt.title(f"Vega vs Time to Maturity (American {option_type.capitalize()}, Binomial)")
    plt.grid(True); plt.show()

   
    r_grid = np.linspace(0.0, 0.10, 30)
    opt = []
    for rr in r_grid:
        V = binomial_option_crr(S0, K, T, rr, sigma, N_sens, option_type=option_type, american=True, q=q)
        opt.append(V)

    plt.figure(figsize=(7,4)); plt.plot(r_grid, opt)
    plt.xlabel("Risk-free rate (r)"); plt.ylabel(f"{option_type.capitalize()} Price")
    plt.title(f"American {option_type.capitalize()} Price vs Interest Rate (Binomial)")
    plt.grid(True); plt.show()
```

```python
sensitivity_plots_american("put")
```
